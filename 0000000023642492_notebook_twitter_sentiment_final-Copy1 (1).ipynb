```python
from nltk.corpus import stopwords
import string
stop_words=stopwords.words('english')
puncuation = list(string.punctuation)
stop_words=stop_words+puncuation
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
        return wordnet.NOUN
```

#r

```python
def clean_review(words):
    output=[]
    for w in words:
        if w.lower() not in stop_words:
            pos=pos_tag([w])
            clean_word= lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output.append(clean_word.lower())
    return output
```

```python
documents=[(clean_review(doc),category) for doc,category in document]
documents[0]
```

```python
doc_test_clean=[clean_review(doc) for doc in doc_test]
```

```python
doc_test_clean[0]
```

```python
sentiment=[category for doc,category in documents]
```

```python
y_train=sentiment
```

```python
x_train = [" ".join(doc) for doc,category in documents]
```

```python
x_test = [" ".join(doc) for doc in doc_test_clean]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect=CountVectorizer(max_features=3000,ngram_range=(1,2))
```

```python
train_data = count_vect.fit_transform(x_train)
```

```python
count_vect.get_feature_names()
```

```python
test_data=count_vect.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(C=260)
```

```python
svc.fit(train_data,y_train)
```

```python
svc.score(train_data,y_train)
```

```python

y_pred=svc.predict(test_data)
y_pred 
```
